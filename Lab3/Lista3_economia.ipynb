```python
import numpy as np
import matplotlib.pyplot as plt
```

### Custo de produção do farmer:
    * Carne: 1 onça/h
    * Batata: 4 onças/h


```python
#x:
carne = np.array([0, 8])
#y:
batata = np.array([32, 0])
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(7, 5))
plt.title('FPP do farmer')
plt.xlabel('Produção de carne(onças)')
plt.ylabel('Produção de batata(onças)')
plt.plot(carne, batata)
plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(7, 5))
plt.title('FPP do farmer')
plt.xlabel('Produção de carne(onças)')
plt.ylabel('Produção de batata(onças)')
plt.plot(carne, batata)
plt.plot(5, 18, 'o')
plt.text(5.1, 18.4, 'A(5,18)')
plt.tight_layout()
plt.show()
```
